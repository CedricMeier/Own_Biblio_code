```python
Donnees_GAGN = pd.read_excel("GAGN_Volume.xlsx",)
#Donnees_GAGN = Donnees_GAGN.replace('-', np.nan)
#Donnees_GAGN = Donnees_GAGN.replace('[<>-]', '', regex=True)

#Transform Date data into datetime
Donnees_GAGN['Date'] = pd.to_datetime(Donnees_GAGN['Date'])

Donnees_GAGN['Date']
Conso=Donnees_GAGN['Torny-conso']
Fourniture=Donnees_GAGN['Torny-fourniture']

#Grouped bar chart with labels
#This example shows a how to create a grouped bar chart and how to annotate bars with labels.

# data from https://allisonhorst.github.io/palmerpenguins/

import matplotlib.pyplot as plt
import numpy as np

Data = ("Dec-22", "jan-23", "feb-23", "mar-23","Apr-23", "May-23", "Jun-23", "July-23", "Aug-23", "Sep-23", "Oct-23", "Nov-23")
Data_means = {
    'Torny-le-Grand Consomation': (Conso),
    'Torny-le-Grand Fourniture': (Fourniture),
}

x = np.arange(len(Data))  # the label locations
width = 0.25  # the width of the bars
multiplier = 1

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in Data_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3, fontsize='small')
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Volume (m3)')
ax.set_title('Volume mensuel')
ax.set_xticks(x + width, Data, rotation='vertical')
#ax.set_xticklabels(rotation=90, fontsize='x-small')#
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(6000, 17000)

plt.savefig('barchart.png')
plt.show()

#The use of the following functions, methods, classes and modules is shown in this example:

#matplotlib.axes.Axes.bar / matplotlib.pyplot.bar

#matplotlib.axes.Axes.bar_label / matplotlib.pyplot.bar_label

```
